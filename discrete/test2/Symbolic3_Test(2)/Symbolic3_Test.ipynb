````python
#Question 1

def u (n):
    sum = Rational(0)
    for k in range (1, n+1):
        sum += Rational(1, k * (k+1))
    return sum

for n in range (5,35,5):
    print(u(n))

for n in range(20):
    print(u(n))


````

<div markdown=1 class="DoIt"> 
We consider the following linear differential equation:
$$
    y''(x)-xy'(x)-y(x)=0.
